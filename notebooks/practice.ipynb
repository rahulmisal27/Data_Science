```
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
data=load_diabetes()
```

```
Features=data.data
Target=data.target
```

```
Features.shape
```

```
Target.shape
```

```
features=pd.DataFrame(Features,columns=data.feature_names)
target=pd.DataFrame(Target)
```

```
x_train=features[:400]
y_train=target[:400]
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```
x_test=features[400:]
```

```
predict=model.predict(x_test)
```

```
y_test=target[400:]
```

```
predict.shape
```

```
y_test.shape
```

```
print("Coeff \n",model.coef_)
```

```
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error,
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, predict))
```

```
print("mean_absolute_error:%.2f"%mean_absolute_error(y_test, predict))
```

```
from sklearn.datasets import california_housing
data=california_housing.fetch_california_housing()
```

```
Features=data.data
Target=data.target
features=pd.DataFrame(Features,columns=data.feature_names)

target=pd.DataFrame(Target)
```

```
features.shape
```

```
target.shape
```

```
x_train=features[:15000]
x_test=features[15000:]
y_train=target[:15000]
y_test=target[15000:]
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
predict=model.predict(x_test)
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error
print("mean_squared_error:%.2f"%mean_squared_error(y_test,predict))
print("mean_absolute_error:%.2f"%mean_absolute_error(y_test,predict))
```
