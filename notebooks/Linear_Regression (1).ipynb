```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
from sklearn.datasets import california_housing
```

```
data=pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```
data.head()
```

```
data.columns
```

```
target=data['median_house_value']
features=data[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income']]
```

```
from sklearn.preprocessing import StandardScaler
StandardScaler(features,target)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.33)
```

```
x_train.shape
```

```
x_test.shape
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```
predict=model.predict(x_test)
```

```
y_test
```

```
from sklearn.metrics import roc_curve, mean_squared_error, mean_absolute_error
```

```
mean_squared_error(predict,y_test)
```

```
mean_absolute_error(predict,y_test)
```
