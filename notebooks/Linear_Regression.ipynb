```
data1=data.drop(['longitude','latitude'],axis=1)
```

```
import seaborn as sns
sns.pairplot(data=data1)
```

```
sns.heatmap(data=data1)
```

```
from scipy import stats
from scipy.stats import skew,norm
from scipy.stats.stats import pearsonr
```

```


# kernel density plot
sns.distplot(data1.median_house_value,fit=norm);
plt.ylabel =('Frequency')
plt.title = ('SalePrice Distribution');
#Get the fitted parameters used by the function
(mu, sigma) = norm.fit(data1['median_house_value']);
#QQ plot
fig = plt.figure()
res = stats.probplot(data1['median_house_value'], plot=plt)
plt.show()
print("skewness: %f" % data1['median_house_value'].skew())
print("kurtosis: %f" % data1['median_house_value'].kurt())

```

```
data1['median_house_value']=np.log1p(data1['median_house_value'])

# kernel density plot
sns.distplot(data1.median_house_value,fit=norm);
plt.ylabel =('Frequency')
plt.title = ('SalePrice Distribution');
#Get the fitted parameters used by the function
(mu, sigma) = norm.fit(data1['median_house_value']);
#QQ plot
fig = plt.figure()
res = stats.probplot(data1['median_house_value'], plot=plt)
plt.show()
print("skewness: %f" % data1['median_house_value'].skew())
print("kurtosis: %f" % data1['median_house_value'].kurt())
```
